```python
import requests
import csv
import time

USERNAME = "kMq26SYmRY"
URL = "https://leetcode.com/graphql"
HEADERS = {
    "Content-Type": "application/json",
    "Referer": f"https://leetcode.com/{USERNAME}/",
    "User-Agent": "Mozilla/5.0"
}

QUERY = """
query Submissions($username: String!, $offset: Int!, $limit: Int!) {
  submissionList(username: $username, offset: $offset, limit: $limit) {
    submissions {
      id
      title
      titleSlug
      statusDisplay
      lang
      timestamp
    }
    hasNext
  }
}
"""

def fetch_submissions(username):
    offset = 0
    limit = 20
    all_submissions = []

    while True:
        variables = {"username": username, "offset": offset, "limit": limit}
        resp = requests.post(URL, json={"query": QUERY, "variables": variables}, headers=HEADERS)
        data = resp.json().get("data", {}).get("submissionList", {})
        subs = data.get("submissions", [])
        has_next = data.get("hasNext", False)

        if not subs:
            break

        all_submissions.extend(subs)
        if not has_next:
            break

        offset += limit
        time.sleep(0.3)

    return all_submissions

def save_to_csv(subs, filename="leetcode_submissions.csv"):
    with open(filename, "w", newline="", encoding="utf-8") as f:
        writer = csv.writer(f)
        writer.writerow(["ID", "Title", "URL", "Status", "Language", "Datetime"])
        for s in subs:
            dt = time.strftime('%Y-%m-%d %H:%M:%S', time.localtime(int(s["timestamp"])))
            writer.writerow([
                s["id"],
                s["title"],
                f"https://leetcode.com/problems/{s['titleSlug']}/",
                s["statusDisplay"],
                s["lang"],
                dt
            ])

# Run process
subs = fetch_submissions(USERNAME)
save_to_csv(subs)
print(f"âœ… {len(subs)} submissions saved!")
```

```python
import requests
import csv
import time
from datetime import datetime

# Step 1: LeetCode API Endpoint and GraphQL Query
USERNAME = "kMq26SYmRY"  # replace with your username
URL = "https://leetcode.com/graphql"

QUERY = """
query recentAcSubmissions($username: String!, $limit: Int!) {
  recentAcSubmissionList(username: $username, limit: $limit) {
    id
    title
    titleSlug
    timestamp
  }
}
"""

# Step 2: Set variables for the query
variables = {
    "username": USERNAME,
    "limit": 99  # Max allowed value
}

# Step 3: Make the POST request
response = requests.post(URL, json={"query": QUERY, "variables": variables})

# Step 4: Parse the response JSON
if response.status_code == 200:
    submissions = response.json()["data"]["recentAcSubmissionList"]
else:
    print("Failed to fetch data:", response.status_code)
    exit()

# Step 5: Prepare CSV file
filename = f"{USERNAME}_leetcode_submissions.csv"

# Step 6: Write to CSV file
with open(filename, mode='w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    # Write header
    writer.writerow(["ID", "Title", "Slug URL", "Submission Time"])

    # Write each submission
    for sub in submissions:
        sub_id = sub["id"]
        title = sub["title"]
        slug_url = f"https://leetcode.com/problems/{sub['titleSlug']}/"
        timestamp = datetime.fromtimestamp(int(sub["timestamp"])).strftime('%Y-%m-%d %H:%M:%S')
        writer.writerow([sub_id, title, slug_url, timestamp])

print(f"âœ… Submissions saved to {filename}")
```
